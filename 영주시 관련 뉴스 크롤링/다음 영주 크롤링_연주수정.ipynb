```python
from selenium import webdriver
from selenium.webdriver.common.by import By #클릭을 하기위한 요소를 찾는 기능
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
import re
pd.set_option('display.max_colwidth', None)
```

```python
query = "영주+비+피해"
news_list = []
for page in range(1, 6):
    url = f"https://search.daum.net/search?nil_suggest=btn&w=news&DA=STC&cluster=y&q={query}&p={page}&sort=old"    
    html_news = requests.get(url).text
    soup_news = BeautifulSoup(html_news, 'html.parser')
    tag_list = soup_news.select('a')
    
    for i in range(0, len(tag_list)):
        info = tag_list[i]['href']
        if 'daum' in info:
            news_list.append(info)
        else:
            continue
    
    time.sleep(1)

filtered_links = set()
pattern = r'http://v\.daum\.net/v/\d+'

for link in news_list:
    match = re.match(pattern, link)
    if match:
        filtered_links.add(match.group())

    news_title = []
    news_contend = []
    news_date = []
```

```python
for url in filtered_links :
    html_news = requests.get(url, headers={'User-Agent' : 'Mozilla/5.0'}).text
    soup_news = BeautifulSoup(html_news, 'html.parser')

    if soup_news.select('h3.tit_view') != [] :
        date = soup_news.select_one('span.num_date')
        news_date.append(date)
    
        title = soup_news.find('h3').get_text()
        news_title.append(title)

        contend = soup_news.select_one('div.article_view section p')
        news_contend.append(contend)
        
news_table = pd.DataFrame([news_date, news_title, news_contend, filtered_links]).T
news_table.columns = ['일자', '뉴스제목', '뉴스내용', '뉴스링크']

```
