```python
df_temp = df[['품종명', '등급명', '총거래물량(kg)', '평균가격(원)', '총거래금액(원)']]
```

```python
df_temp
```

```python
df_temp.isnull().sum()
```

```python
X = df_temp[['평균가격(원)', '총거래물량(kg)', '총거래금액(원)']].values
y = df_temp['등급명'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
```

```python
from sklearn.ensemble import RandomForestClassifier

# 로지스틱 회귀
lr_clf = LogisticRegression(max_iter = 100000)

# 랜포
rf_clf = RandomForestClassifier(n_jobs=-1, random_state=42)
rf_clf.fit(X_train, y_train)

# KNN
knn_clf = KNeighborsClassifier(n_neighbors=18)

# 결정트리
dt_clf = DecisionTreeClassifier(max_depth= 3)
```

```python
lr_clf.fit(X_train, y_train) 
print(f'train score: {lr_clf.score(X_train, y_train)}')
print(f'test score: {lr_clf.score(X_test, y_test)}')
```

```python
rf_clf.fit(X_train, y_train)
print(f'train score: {rf_clf.score(X_train, y_train)}')
print(f'test score: {rf_clf.score(X_test, y_test)}')
```

```python
knn_clf.fit(X_train, y_train)

print(f'train score: {knn_clf.score(X_train, y_train)}')
print(f'test score: {knn_clf.score(X_test, y_test)}')
```

```python
dt_clf.fit(X_train, y_train)

print(f'train score: {dt_clf.score(X_train, y_train)}')
print(f'test score: {dt_clf.score(X_test, y_test)}')
```

```python
vo_clf = VotingClassifier(estimators=[('RF', rf_clf),('LR' , lr_clf), ('KNN', knn_clf), ('DT', dt_clf)], voting= 'soft')
```

```python
vo_clf.fit(X_train, y_train)
```

```python
y_pred_test = vo_clf.predict(X_test)
test_score = accuracy_score(y_test, y_pred_test)
print(f'voting test score: {test_score}')
```

```python
clf_list = [rf_clf,lr_clf, knn_clf, dt_clf]

for clf in clf_list:
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

```python
# 에이다부스트 
from sklearn.ensemble import AdaBoostClassifier
ada_clf = AdaBoostClassifier(n_estimators=50, algorithm='SAMME.R', learning_rate=1)
ada_clf.fit(X_train, y_train)

print(f"train_score: {ada_clf.score(X_train, y_train)}")
print(f"test_score: {ada_clf.score(X_test, y_test)}")
```

```python
# 그래디언트부스트
from sklearn.ensemble import GradientBoostingClassifier
basic_gb = GradientBoostingClassifier(random_state=42)

basic_gb.fit(X_train, y_train)

print(f"train: {basic_gb.score(X_train, y_train)}")
print(f"test_score: {basic_gb.score(X_test, y_test)}")
```

# 아래는 성능 높이기 gridsearch

```python
params = {
    'n_estimators': [10, 50, 100, 300, 500],
    'learning_rate':[0.001, 0.01, 0.1, 1, 10, 100],
    'max_depth': [1, 2, 4, 8],
    'min_samples_split':[2, 4,8,16]
}

basic_gb = GradientBoostingClassifier(random_state=42)
grid = GridSearchCV(basic_gb,\
             param_grid=params, n_jobs=-1, cv=5, scoring='accuracy')

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
print(f"train: {grid.best_estimator_.score(X_train, y_train)}")
print(f"test_score: {grid.best_estimator_.score(X_test, y_test)}")
```
