```python
import squarify
import matplotlib.pyplot as plt
import random

# 생산량을 기준으로 내림차순으로 데이터 프레임 정렬
df.sort_values('생산량', ascending=False, inplace=True)

# 각 지역에 대해 색상 설정
colors = ['#FFCDEA' if region == '영주시' else random.choice([ '#C5EBAA', '#DDDDDD', '#EEEEEE', '#E2F4C5', '#58A399', '#A8CD9F']) for region in df['지역']]

fig, ax = plt.subplots(1, figsize=(10,4))
squarify.plot(sizes=df['생산량'], label=df['지역'], color=colors, alpha=0.8)
plt.axis('off')
plt.title('경북 사과 생산량')
plt.show()
```
